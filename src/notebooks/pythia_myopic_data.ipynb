```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
sizes = np.array([
    1.4e7, 3.1e7, 7.0e7, 1.6e8, 4.1e8, 1.0e9, 
    1.4e9, 2.8e9
])

vanilla = np.array([
    4.11, 3.73, 3.44, 3.03, 2.74, 2.62, 2.54, 2.44
])

myopic = np.array([
    4.27, 3.84, 3.48, 3.13, 2.89, 2.78, 2.71, 2.67
])


# OLD -- incorrect precision
# vanilla = np.array([
#     4.28, 3.94, 3.65, 3.36, 2.97, #2.85, 
#     2.80, #2.73
# ])
# myopic = np.array([
#     4.39, 4.03, 3.72, 3.48, 3.13, #3.00, 
#     2.99, #3.23
# ])
```

```python
plt.plot(sizes, vanilla, label='vanilla')
plt.plot(sizes, myopic, label='myopic')
plt.xscale('log')
plt.xlabel('# parameters')
plt.ylabel('cross-entropy loss')
plt.legend()
plt.show()
```

```python
plt.plot(sizes, myopic - vanilla)
plt.xscale('log')
plt.xlabel('# parameters')
plt.ylabel('myopia gap')
plt.show()
```
